```python
import pandas
```

```python
research = pandas.read_csv("PhysFit1999.txt",dtype=str,index_col=False)
```

```python
research.rename(columns = {'Ccode': 'ccode', 'Dcode':'dcode', 'Scode':'scode', 'charternum':'chrtnum'}, inplace = True)
```

```python
research
```

```python
oc_research = research.query("`scode` == '0000000' and `dcode` != '00000' and `ccode` == '30' and `Level` == '2' and `RptType` == '1' and `SubGrp` == '0' and (`Line_num` == '1' or `Line_num` == '2')")
```

```python
oc_research
```

```python
def weighted_average(df, values, weights):
    d = df[values].astype(float)
    w = df[weights].astype(int)
    return (d * w).sum() / w.sum()
```

```python
oc_research.is_copy = False
oc_research.loc[oc_research["Gr5PctIn"] == '*', "Gr5PctIn"] = '0.00 '
oc_research.loc[oc_research["Gr7PctIn"] == '*', "Gr7PctIn"] = '0.00 '
oc_research.loc[oc_research["Gr9PctIn"] == '*', "Gr9PctIn"] = '0.00 '
weighted_5 = str(weighted_average(oc_research, 'Gr5PctIn', 'Gr05_Stu'))
weighted_5
```

```python
oc_research.loc[oc_research["Gr5PctIn"] == '0.00 ', "Gr5PctIn"] = weighted_5
oc_research
```

```python
weighted_7 = str(weighted_average(oc_research, 'Gr7PctIn', 'Gr07_Stu'))
oc_research.loc[oc_research["Gr7PctIn"] == '0.00 ', "Gr7PctIn"] = weighted_7
weighted_7
```

```python
weighted_9 = str(weighted_average(oc_research, 'Gr9PctIn', 'Gr09_Stu'))
oc_research.loc[oc_research["Gr9PctIn"] == '0.00 ', "Gr9PctIn"] = weighted_9
weighted_9
```

```python
oc_research.to_csv('oc_result99.csv')
oc_research
```
